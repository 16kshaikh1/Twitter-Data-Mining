```python
#Get reciprocal friends
friends = my_friends()
followers = my_followers()

#Distance 1 friends
def reciprocal_friends():
    rec_friends = []
    for x in friends:
        for y in followers:
            if(x==y):
                rec_friends.append(x)
    return rec_friends
reciprocal_friends()      
```
